```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<center><b>Завдання 7.2. Визначити розмір зображення та зсунути зображення на N*20 стовпців та N*10 рядків, де N - номер варіанту.<b><center>

```python
img = cv2.imread(r'E:\KHPI\Computer Vision\Lab_7\pics\12.jpg')
rows= img.shape[0]
cols = img.shape[1]
M = np.float32([[1,0,240],[0,1,120]])

dst = cv2.warpAffine(img, M, (cols,rows))

cv2.imshow('Changed image', dst)
cv2.waitKey(0)
```

<center><b>Завдання 7.3. Визначити розмір зображення, його центр та повернути його на Nx10 градусів (N- номер варіанту)<b><center>

```python
img = cv2.imread(r'E:\KHPI\Computer Vision\Lab_7\pics\12.jpg')
rows = img.shape[0]
cols = img.shape[1]

M = cv2.getRotationMatrix2D((cols / 2, rows / 2), 120, 1)
dst = cv2.warpAffine(img, M, (cols, rows))

cv2.imshow('Changed image', dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<center><b>Завдання 7.4. Визначити розмір зображення, задати 3 точки, змінити їх координати та провести афінне перетворення всього зображення за цими точками.<b><center>

```python
img = cv2.imread(r'E:\KHPI\Computer Vision\Lab_7\pics\12.jpg')
cv2.imshow("Original", img)

rows = img.shape[0]
cols = img.shape[1]

pts1= np.float32([[50,50],[200,50],[50,200]])
pts2= np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)
dst= cv2.warpAffine(img, M, (cols,rows))

#transforming color from rgb -> bgr
bgr_image = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
bgr_dst = cv2.cvtColor(dst, cv2.COLOR_RGB2BGR)

plt.subplot(121), plt.imshow(bgr_image), plt.title('Input')
plt.subplot(122), plt.imshow(bgr_dst), plt.title('Output')

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<center><b>Завдання 7.5. Намалювати прямокутник у місці, де потрібно вирізати фрагмент (номерний знак), вивести на екран фрагмент, обмежений прямокутником, збільшивши цей фрагмент. Визначити розмір зображення, його центр та повернути його на 90 градусів.<b><center>

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread(r'E:\KHPI\Computer Vision\Lab_7\pics\12.jpg')
image = cv2.rectangle(img, (1100, 731), (1348, 644), (0, 0, 255), 2)
cv2.imshow("Image", image)

#y1y2, x1x2
crop = img[650:730, 1100:1350]
#cv2.imshow("Area", crop)

piece = cv2.resize(crop,  (200,50), interpolation=cv2.INTER_LINEAR)
cv2.imshow("Area", piece)

(h, w) = piece.shape[:2]
print(w,h)
center= (w / 2, h / 2)

M = cv2.getRotationMatrix2D(center,  90, 2.0)
rotated = cv2.warpAffine(piece, M, (150, 200))

cv2.imshow("Rotated", rotated)
cv2.waitKey()
cv2.destroyAllWindows()
```
