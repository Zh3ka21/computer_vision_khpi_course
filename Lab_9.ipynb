```python
import cv2
import numpy as np
```

```python
path_img = r'E:\KHPI\Computer Vision\Lab_9\pics\12.jpg'
save_path = r'E:\KHPI\Computer Vision\Lab_9\pics'
```

```python
img = cv2.imread(r'E:\KHPI\Computer Vision\Lab_9\pics\12.jpg', 0)
plate = img[650:730, 1100:1350]

cv2.imshow('image', plate)
cv2.imwrite(f"{save_path}license_plate.jpg", plate)

ret, thresh = cv2.threshold(plate,  180, 255,  cv2.THRESH_BINARY)
contours, hierarchy = cv2.findContours(thresh,5, 5)

print(cv2.findContours(thresh,5, 5))

contour = contours[0]


contourArea = cv2.contourArea(contour)
print(f"Площа контура = {contourArea}")

contourLength = cv2.arcLength(contour, True)
print(f"Периметр контура = {contourLength}")

x, y, w, h = cv2.boundingRect(contour)
print(f"W = {w}, H = {h}, W/H = {round(w/h,3)}")

area_ratio = (plate.shape[0] * plate.shape[1]) / (w * h)
print(f"Area ratio = {area_ratio}")

eqdiam = np.sqrt(4 * contourArea/np.pi)
print(f"Єквівалентний діаметр = {eqdiam}")

moment = cv2.moments(contour)
print(f"Моменти = {moment['m00']}, {moment['m01']}, {moment['m10']}")

cx = int(moment['m10']/moment['m00'])
cy = int(moment['m01']/moment['m00'])
print(f"Cx = {cx}, Cy = {cy}")

mask = np.zeros(plate.shape, np.uint8)
minv, maxv, minl, maxl = cv2.minMaxLoc(plate, mask=mask)
print(f"Мінімальне та максимальне значення  = {minv, maxv, minl, maxl}")

leftmost = tuple(contour[contour[:, :, 0].argmin()][0])
rightmost = tuple(contour[contour[:, :, 0].argmax()][0])
topmost = tuple(contour[contour[:, :, 1].argmin()][0])
bottommost = tuple(contour[contour[:, :, 1].argmax()][0])
print(f"Крайні точки:\nЛіва = {leftmost}\nПрава = {rightmost}\nВерхня = {topmost}\nНижня = {bottommost}")

(x, y), (MA, ma), ang = cv2.fitEllipse(contour)
print(f"Орієнтація = {ang}")

meanv = cv2.mean(plate, mask=mask)
print(f"Інтенсивність = {meanv}")


cv2.imshow("bin", thresh)
bim_img = np.ones(img.shape, dtype=np.uint8) * 255
cv2.drawContours(bim_img, contour, -1, (0, 255, 0), 3)
cv2.circle(bim_img, (cx, cy), 2, (0, 255, 0), 2)
cv2.imshow("contour", bim_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
